```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"  # embedding তৈরির সময় যেটা ব্যবহার করেছিলে
)

db = Chroma(
    persist_directory="./chroma_db",
    embedding_function=embeddings
)
```

কতগুলো Chunk Store হয়েছে?

```python
collection = db._collection

print("Total Chunks:", collection.count())
```

প্রথম 5টা Chunk দেখো

```python
data = collection.get()

for i, doc in enumerate(data["documents"][:5]):
    print(f"\nChunk {i+1}")
    print(doc[:500])
```

Stored Metadata দেখো

```python
data = collection.get()

for meta in data["metadatas"][:5]:
    print(meta)
```

Similarity Search Test

```python
results = db.similarity_search(
    "What is Abu Sufiun?",
    k=3
)

for i, doc in enumerate(results):
    print(f"\nResult {i+1}")
    print(doc.page_content[:500])
```

Similarity Score সহ Search

```python
results = db.similarity_search_with_score(
    "Abu Sufiun?",
    k=3
)

for doc, score in results:
    print("\nScore:", score)
    print(doc.page_content[:300])
```
